I wanted to make scatter plots to show if pollution was increasing or decreasing

```python
import matplotlib.pyplot as plt
import seaborn as sns

california_pollution_city_counts = california_pollution['City'].value_counts().reset_index()
california_pollution_city_counts.columns = ['City', 'Pollution_Count']

traffic_city_counts = traffic['CITY'].value_counts().reset_index()
traffic_city_counts.columns = ['City', 'Traffic_Count']

combined_city_counts = pd.merge(
    california_pollution_city_counts,
    traffic_city_counts,
    on='City',
    how='outer'
).fillna(0)

combined_city_counts['Pollution_Count'] = combined_city_counts['Pollution_Count'].astype(int)
combined_city_counts['Traffic_Count'] = combined_city_counts['Traffic_Count'].astype(int)

combined_city_counts['Total_Count'] = combined_city_counts['Pollution_Count'] + combined_city_counts['Traffic_Count']
combined_city_counts = combined_city_counts.sort_values(by='Total_Count', ascending=False)

melted_counts = combined_city_counts.melt(
    id_vars=['City'],
    value_vars=['Pollution_Count', 'Traffic_Count'],
    var_name='Dataset',
    value_name='Count'
)

plt.figure(figsize=(16, 8))
sns.barplot(x='City', y='Count', hue='Dataset', data=melted_counts, palette={'Pollution_Count': 'skyblue', 'Traffic_Count': 'lightcoral'})
plt.title('Comparison of City Counts in Pollution and Traffic Datasets')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right') # Rotate labels for better readability
plt.legend(title='Dataset')
plt.tight_layout()
plt.show()
```

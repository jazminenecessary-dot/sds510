```python
import pandas as pd
import networkx as nx
```

```python
siren_df = pd.read_csv('SIREN.csv', index_col=0)
siren_df.columns = siren_df.index # Ensure columns match index for adjacency matrix
s = nx.from_pandas_adjacency(siren_df.fillna(0))
nx.draw_networkx(s) # Draw the graph 's' created from siren_df
```

We can see with the graph that 40 and 42 are futher away from the others.

```python
print(f"Number of nodes: {s.number_of_nodes()}")
print(f"Number of edges: {s.number_of_edges()}")
```

```python
nx.average_clustering(s)
```

```python
nx.betweenness_centrality(s)
```
